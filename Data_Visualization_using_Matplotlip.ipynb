```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#Reading Dataset 
movie_df = pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/boston_train.csv', sep = ',')
```

```
display(movie_df)
```

```
movie_df.isnull().sum()
```

Line Plot 
Use line plot for Univarairte Data Anaylisis,Continious




```
plt.figure(figsize =(20,9))
plt.plot(movie_df['AGE'])
plt.title("AGE Line Plot",fontsize = 14)
plt.xlabel("Row Indexes")
plt.ylabel("AGE")
plt.show()
```

Bar Plot 


*   Categorical / Discrete values vs Y could be contineous or Disscrete 




```
plt.figure(figsize = (12,8))
plt.bar(x = movie_df['ZN'], height= movie_df['AGE'], width = 1.5, color = 'turquoise')
plt.title("Bar Plot", fontsize = 14)
plt.xlabel("ZN")
plt.ylabel("Max of AGE")
plt.show()
```

Histogram Plot 


*   Univerate Analysis for single values 
*   Variance in Distribution of dataset



```
..
1 - 60
[4,5,9,11,12,15,21,23,34,41,44,43,55,51]
```

```
plt.figure(figsize=(12,8))
plt.hist(movie_df['MEDV'],bins=20)
plt.title("Histogram Plot",fontsize = 14)
plt.xlabel("MEDV")
plt.ylabel("Frequency (in Range)")
plt.show()
```

Scatter Plot 




```
iris_df = pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/iris.csv', sep = ',')
iris_df
```

```
iris_df['species'].unique()
```

```
plt.figure(figsize=(12,8))
plt.scatter(x=iris_df['sepal_length'],y=iris_df['petal_length'])
plt.title("Scatter Plot", fontsize = 14)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```
plt.figure(figsize = (12,8))
sns.scatterplot( x = 'sepal_length', y = 'petal_length', data = iris_df, hue = 'species', palette='magma')
plt.show()
```


pie chart


```
plt.figure(figsize=(10,10))
plt.pie(iris_df['species'].value_counts(),autopct='%0.2f%%',labels = ['setosa', 'versicolor', 'verginica'],explode= (0, 0, 0.2))
plt.legend()
plt.show()
```

Distribution Plot

* Density plot , Histogram plot



```
plt.figure(figsize=(11,8))
sns.distplot(movie_df['MEDV'], color = 'darkviolet')
plt.show()

```
